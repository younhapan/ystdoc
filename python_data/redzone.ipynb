```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://thelogicalindian.com/news/districts-red-orange-green-zones-covid-19-lockdown-20882'
c = requests.get(url).text
soup = BeautifulSoup(c)
l = soup.find_all('div', class_='story-data')[0]
p_re = re.compile('^\s*[0-9]')
```

```python
for ele in l.find_all('p'):
    text = ele.contents[0]
    try:
        if p_re.search(text):
            print(text)
    except:
        continue
```
